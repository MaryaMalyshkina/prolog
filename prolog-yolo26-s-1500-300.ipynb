```python
import torch
import pandas as pd
import numpy as np
import os
import cv2
import matplotlib.pyplot as plt
from PIL import Image
from matplotlib.figure import Figure
from pathlib import Path
import random
import shutil
```

```python
from ultralytics import YOLO

# Load a pre-trained YOLO11 instance segmentation model
model = YOLO("yolo26s-seg.pt")
```

```python
# ĞŸÑ€Ğ¾Ğ²ĞµÑ€ĞºĞ° Ğ´Ğ¾ÑÑ‚ÑƒĞ¿Ğ½Ğ¾ÑÑ‚Ğ¸ GPU
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")
```

```python
results = model.train(
    data="/kaggle/input/prolog-1500/svik54.v3-prolog_1500.yolov11/data.yaml",
    epochs=300,

    imgsz=640,
    batch=16,
    workers=2,

    # Ğ“Ğ¸Ğ¿ĞµÑ€Ğ¿Ğ°Ñ€Ğ°Ğ¼ĞµÑ‚Ñ€Ñ‹
    lr0=0.01,                 # Ğ½Ğ°Ñ‡Ğ°Ğ»ÑŒĞ½Ñ‹Ğ¹ learning rate
    lrf=0.001,                 # ĞºĞ¾Ğ½ĞµÑ‡Ğ½Ñ‹Ğ¹ learning rate
    momentum=0.937,           # Ğ¼Ğ¾Ğ¼ĞµĞ½Ñ‚ SGD
    weight_decay=0.0005,      # Ğ²ĞµÑĞ¾Ğ²Ğ°Ñ Ñ€ĞµĞ³ÑƒĞ»ÑÑ€Ğ¸Ğ·Ğ°Ñ†Ğ¸Ñ
    warmup_epochs=3,          # Ğ¿Ñ€Ğ¾Ğ³Ñ€ĞµĞ²
    warmup_momentum=0.8,      # Ğ¼Ğ¾Ğ¼ĞµĞ½Ñ‚ Ğ¿Ñ€Ğ¸ Ğ¿Ñ€Ğ¾Ğ³Ñ€ĞµĞ²Ğµ
    warmup_bias_lr=0.1,       # lr Ğ´Ğ»Ñ bias Ğ¿Ñ€Ğ¸ Ğ¿Ñ€Ğ¾Ğ³Ñ€ĞµĞ²Ğµ


    patience=50,              # Ñ€Ğ°Ğ½Ğ½ÑÑ Ğ¾ÑÑ‚Ğ°Ğ½Ğ¾Ğ²ĞºĞ° ĞµÑĞ»Ğ¸ Ğ½ĞµÑ‚ ÑƒĞ»ÑƒÑ‡ÑˆĞµĞ½Ğ¸Ğ¹

    # ĞĞ¿Ñ‚Ğ¸Ğ¼Ğ¸Ğ·Ğ°Ñ†Ğ¸Ñ
    optimizer="SGD",          # Ğ¼Ğ¾Ğ¶Ğ½Ğ¾ "AdamW" Ğ´Ğ»Ñ Ğ¼Ğ°Ğ»ĞµĞ½ÑŒĞºĞ¾Ğ³Ğ¾ Ğ´Ğ°Ñ‚Ğ°ÑĞµÑ‚Ğ°
    cos_lr=True,              # ĞºĞ¾ÑĞ¸Ğ½ÑƒÑĞ½Ñ‹Ğ¹ scheduler
    close_mosaic=10,          # Ğ¾Ñ‚ĞºĞ»ÑÑ‡Ğ¸Ñ‚ÑŒ mosaic Ğ·Ğ° 10 ÑĞ¿Ğ¾Ñ… Ğ´Ğ¾ ĞºĞ¾Ğ½Ñ†Ğ°

    # Ğ’Ğ°Ğ»Ğ¸Ğ´Ğ°Ñ†Ğ¸Ñ
    val=True,                 # Ğ²Ğ°Ğ»Ğ¸Ğ´Ğ°Ñ†Ğ¸Ñ Ğ¿Ğ¾ÑĞ»Ğµ ĞºĞ°Ğ¶Ğ´Ğ¾Ğ¹ ÑĞ¿Ğ¾Ñ…Ğ¸
    save_period=10,           # ÑĞ¾Ñ…Ñ€Ğ°Ğ½ÑÑ‚ÑŒ Ñ‡ĞµĞºĞ¿Ğ¾Ğ¸Ğ½Ñ‚Ñ‹ ĞºĞ°Ğ¶Ğ´Ñ‹Ğµ 10 ÑĞ¿Ğ¾Ñ…

    # Ğ”Ñ€ÑƒĞ³Ğ¾Ğµ
    device="cuda" if torch.cuda.is_available() else "cpu",
    seed=42,                  # Ğ´Ğ»Ñ Ğ²Ğ¾ÑĞ¿Ñ€Ğ¾Ğ¸Ğ·Ğ²Ğ¾Ğ´Ğ¸Ğ¼Ğ¾ÑÑ‚Ğ¸
    deterministic=True,       # Ğ´ĞµÑ‚ĞµÑ€Ğ¼Ğ¸Ğ½Ğ¸Ñ€Ğ¾Ğ²Ğ°Ğ½Ğ½Ñ‹Ğµ Ğ°Ğ»Ğ³Ğ¾Ñ€Ğ¸Ñ‚Ğ¼Ñ‹
    single_cls=False,         # Ğ¼ÑƒĞ»ÑŒÑ‚Ğ¸ĞºĞ»Ğ°ÑÑĞ¾Ğ²Ğ°Ñ ĞºĞ»Ğ°ÑÑĞ¸Ñ„Ğ¸ĞºĞ°Ñ†Ğ¸Ñ
    plots=True,               # ÑÑ‚Ñ€Ğ¾Ğ¸Ñ‚ÑŒ Ğ³Ñ€Ğ°Ñ„Ğ¸ĞºĞ¸
    verbose=True,             # Ğ¿Ğ¾Ğ´Ñ€Ğ¾Ğ±Ğ½Ñ‹Ğ¹ Ğ²Ñ‹Ğ²Ğ¾Ğ´
)

model.save("prolog_s_1500_300_model.pt")
```

```python
# Ğ’Ğ°Ğ»Ğ¸Ğ´Ğ°Ñ†Ğ¸Ñ Ğ¼Ğ¾Ğ´ĞµĞ»Ğ¸
metrics = model.val(data="/kaggle/input/prolog-1500/svik54.v3-prolog_1500.yolov11/data.yaml")

print("ĞœĞµÑ‚Ñ€Ğ¸ĞºĞ¸ Ğ´ĞµÑ‚ĞµĞºÑ†Ğ¸Ğ¸:")
print(f"  mAP50: {metrics.box.map50:.4f}")
print(f"  mAP50-95: {metrics.box.map:.4f}")
print(f"  Precision: {np.mean(metrics.box.p):.4f}")

print(f"  Recall: {np.mean(metrics.box.r):.4f}")

print("\nĞœĞµÑ‚Ñ€Ğ¸ĞºĞ¸ ÑĞµĞ³Ğ¼ĞµĞ½Ñ‚Ğ°Ñ†Ğ¸Ğ¸:")
print(f"  Mask mAP50: {metrics.seg.map50:.4f}")
print(f"  Mask mAP50-95: {metrics.seg.map:.4f}")
```
